### **1.Combine Textual Features**

````python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Merge text fields into a unified text feature column
data_copy['combined_features'] = data_copy['overview'] + ' ' + \
                  data_copy['genres'] + ' ' + \
                  data_copy['keywords']
````

### **2.TF-IDF**

````python
from sklearn.feature_extraction.text import TfidfVectorizer

# Creating a TF-IDF vectorizer
tfidf = TfidfVectorizer(stop_words='english',
                        max_features=5000,
                        ngram_range=(1, 2),
                        min_df=5,
                        max_df=0.9)

# Convert text features into vectors (sparse matrices)
tfidf_matrix = tfidf.fit_transform(data_copy['combined_features'])
````

### **3.Normalize Numerical Features**

````python
from sklearn.preprocessing import MinMaxScaler

# Fill in empty values â€‹â€‹(if any) and scale ratings and box office
numeric_features = data_copy[['vote_average', 'revenue']].fillna(0)

# Normalize (0~1) to facilitate concatenation with text features
scaler = MinMaxScaler()
scaled_numeric = scaler.fit_transform(numeric_features)
````

### **4.Hybrid Features**

````python
from scipy.sparse import hstack, csr_matrix

# Concatenate TF-IDF text matrix and structured numeric matrix
combined_matrix = hstack([tfidf_matrix, csr_matrix(scaled_numeric)])
````

````python
from sklearn.metrics.pairwise import cosine_similarity

# Similarity matrix based on hybrid features
cosine_sim_hybrid = cosine_similarity(combined_matrix, combined_matrix)
````

### **5.Movie Recommendation**

````python
def get_recommendations(title, cosine_sim=cosine_sim_hybrid):
    idx = data_copy[data_copy['title'] == title].index[0]

    # Calculate similarity with all other movies
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sorting: The most similar ones come first
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]  # Return to top 10 similar movies

    # Extract the title of the corresponding movie
    movie_indices = [i[0] for i in sim_scores]
    return data_copy['title'].iloc[movie_indices]
````

###**Feature Engineering for Predictive Modeling**

````python
# Feature Engineering Function
def prepare_features(df):
    """
    Prepare features for predictive modeling
    """
    print(" Preparing features for predictive modeling...")

    df_copy = df.copy()

    # Create target variables
    df_copy['high_rating'] = (df_copy['vote_average'] > df_copy['vote_average'].median()).astype(int)
    df_copy['high_popularity'] = (df_copy['popularity'] > df_copy['popularity'].median()).astype(int)

    # Text feature engineering
    df_copy['overview_length'] = df_copy['overview'].apply(lambda x: len(str(x)))
    df_copy['tagline_length'] = df_copy['tagline'].apply(lambda x: len(str(x)))
    df_copy['cast_count'] = df_copy['cast'].apply(lambda x: len(str(x).split()))

    # Extract year from release date
    df_copy['release_year'] = pd.to_datetime(df_copy['release_date']).dt.year
    df_copy['years_old'] = 2024 - df_copy['release_year']  # Assuming current year is 2024

    # Budget and revenue features
    df_copy['budget_revenue_ratio'] = np.where(
        df_copy['budget'] > 0,
        df_copy['revenue'] / df_copy['budget'],
        0
    )
    df_copy['profit'] = df_copy['revenue'] - df_copy['budget']

    # Genre encoding
    all_genres = set()
    for genres in df_copy['genres']:
        if genres != 'nan' and pd.notna(genres):
            genre_list = str(genres).split()
            all_genres.update(genre_list)

    # Create genre columns
    for genre in all_genres:
        if genre:  # Skip empty strings
            col_name = f'genre_{genre.lower().replace(" ", "_")}'
            df_copy[col_name] = df_copy['genres'].apply(
                lambda x: 1 if genre in str(x) else 0
            )

    # Select features for modeling
    numeric_features = [
        'budget', 'revenue', 'runtime', 'popularity', 'vote_count',
        'overview_length', 'tagline_length', 'cast_count',
        'release_year', 'years_old', 'budget_revenue_ratio', 'profit'
    ]

    # Get genre columns
    genre_columns = [col for col in df_copy.columns if col.startswith('genre_')]

    # Combine all features
    feature_columns = numeric_features + genre_columns

    # Remove any potential NaN values
    df_copy = df_copy.dropna(subset=numeric_features + ['vote_average', 'high_rating', 'high_popularity'])

    print(f" Features prepared: {len(feature_columns)} total features")
    print(f" Final dataset size: {len(df_copy)} movies")

    return df_copy, feature_columns

# Apply feature engineering
df_modeling, feature_columns = prepare_features(data_copy)

# Display feature information
print(f"\n Feature Overview:")
print(f"â€¢ Numeric features: {len([f for f in feature_columns if not f.startswith('genre_')])}")
print(f"â€¢ Genre features: {len([f for f in feature_columns if f.startswith('genre_')])}")
print(f"â€¢ Total features: {len(feature_columns)}")
````

###**Content-Based Recommendation System**

````python
from sklearn.metrics.pairwise import cosine_similarity
from scipy.sparse import hstack, csr_matrix
import pandas as pd
import numpy as np
import re

class ContentBasedRecommender:
    def __init__(self, df):
        self.df = df.reset_index(drop=True)
        self.tfidf_matrix = None
        self.cosine_sim = None
        self.feature_names = None
        self._prepare_similarity_matrix()

    def _prepare_similarity_matrix(self):
        """Prepare TF-IDF matrix and cosine similarity"""
        print(" Preparing content-based similarity matrix...")

        # Combine text features for content analysis
        self.df['content'] = (
            self.df['genres'].fillna('') + ' ' +
            self.df['overview'].fillna('') + ' ' +
            self.df['keywords'].fillna('') + ' ' +
            self.df['cast'].fillna('') + ' ' +
            self.df['director'].fillna('') + ' ' +
            self.df['tagline'].fillna('')
        )

        # Clean text data
        self.df['content'] = self.df['content'].apply(
            lambda x: re.sub(r'[^\w\s]', '', str(x).lower())
        )

        # Create TF-IDF matrix
        tfidf = TfidfVectorizer(
            stop_words='english',
            max_features=5000,
            min_df=2,
            max_df=0.8
        )
        self.tfidf_matrix = tfidf.fit_transform(self.df['content'])
        self.feature_names = tfidf.get_feature_names_out()

        # Compute cosine similarity
        self.cosine_sim = cosine_similarity(self.tfidf_matrix) # Removed redundant self.cosine_sim

        print(f"Content matrix prepared: {self.tfidf_matrix.shape[0]} movies, {self.tfidf_matrix.shape[1]} features")

    def get_recommendations(self, movie_title, top_n=10):
        """Get content-based recommendations"""
        try:
            # Find movie index
            matches = self.df[self.df['title'].str.lower() == movie_title.lower()]
            if len(matches) == 0:
                # Try partial match
                matches = self.df[self.df['title'].str.lower().str.contains(movie_title.lower())]
                if len(matches) == 0:
                    print(f"Movie '{movie_title}' not found in dataset.")
                    return None

            idx = matches.index[0]
            actual_title = self.df.iloc[idx]['title']
            print(f"ðŸŽ¬ Found movie: '{actual_title}'")

            # Get similarity scores
            sim_scores = list(enumerate(self.cosine_sim[idx]))

            # Sort by similarity score
            sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

            # Get top N similar movies (excluding itself)
            sim_scores = sim_scores[1:top_n+1]

            # Get movie indices
            movie_indices = [i[0] for i in sim_scores]

            # Return recommendations
            recommendations = self.df.iloc[movie_indices][['title', 'genres', 'vote_average', 'popularity']].copy()
            recommendations['similarity_score'] = [i[1] for i in sim_scores]
            recommendations['similarity_percentage'] = (recommendations['similarity_score'] * 100).round(2)

            return recommendations

        except Exception as e:
            print(f"Error getting recommendations: {e}")
            return None

    def explain_similarity(self, movie1, movie2):
        """Explain why two movies are similar"""
        try:
            idx1 = self.df[self.df['title'].str.lower() == movie1.lower()].index[0]
            idx2 = self.df[self.df['title'].str.lower() == movie2.lower()].index[0]

            # Get common features with high TF-IDF scores
            features1 = self.tfidf_matrix[idx1].toarray().flatten()
            features2 = self.tfidf_matrix[idx2].toarray().flatten()

            # Find common high-scoring features
            common_indices = np.where((features1 > 0) & (features2 > 0))[0]
            common_features = [(self.feature_names[i], features1[i], features2[i])
                             for i in common_indices if features1[i] > 0.1 or features2[i] > 0.1]

            # Sort by combined score
            common_features.sort(key=lambda x: x[1] + x[2], reverse=True)

            print(f"\nSimilarity explanation between '{movie1}' and '{movie2}':")
            print("Common features contributing to similarity:")
            for feature, score1, score2 in common_features[:10]:  # Top 10 features
                print(f"   â€¢ {feature}: {score1:.3f} + {score2:.3f}")

        except Exception as e:
            print(f"Error explaining similarity: {e}")

# Initialize content-based recommender
print("\n INITIALIZING CONTENT-BASED RECOMMENDER")
content_recommender = ContentBasedRecommender(data_copy)

# Test the recommender with user input
test_movie = input("Enter the movie title you want content-based recommendations for: ")
print(f"\n TESTING CONTENT-BASED RECOMMENDATIONS FOR: '{test_movie}'")
recommendations = content_recommender.get_recommendations(test_movie, top_n=5)

if recommendations is not None:
    print(f"\nTop 5 content-based recommendations for '{test_movie}':")
    print(recommendations[['title', 'genres', 'vote_average', 'similarity_percentage']])

    # Explain similarity for top recommendation
    if len(recommendations) > 0:
        top_rec = recommendations.iloc[0]['title']
        content_recommender.explain_similarity(test_movie, top_rec)
````

````python
from sklearn.metrics.pairwise import cosine_similarity
from scipy.sparse import hstack, csr_matrix
import pandas as pd
import numpy as np
import re

class HybridRecommender:
    def __init__(self, df, content_recommender, rating_model, popularity_model, gemini_interpreter, generative_enricher):
        self.df = df.reset_index(drop=True)
        self.content_recommender = content_recommender
        self.rating_model = rating_model
        self.popularity_model = popularity_model
        # Correctly get feature columns from the DataFrame used for training
        # Assuming feature_columns was defined when creating df_modeling
        self.feature_columns = [col for col in df.columns if col.startswith(('budget', 'revenue', 'runtime', 'popularity', 'vote_count',
                                'overview_length', 'tagline_length', 'cast_count',
                                'release_year', 'years_old', 'budget_revenue_ratio', 'profit', 'genre_'))] # Get numeric and genre features

        self.gemini_interpreter = gemini_interpreter
        self.generative_enricher = generative_enricher


    def get_hybrid_recommendations(self, movie_title, query=None, top_n=10,
                                 content_weight=0.5, rating_weight=0.3, popularity_weight=0.2):
        """
        Get hybrid recommendations combining multiple approaches, with optional LLM interpretation.
        """
        print(f" Generating hybrid recommendations for '{movie_title}' with query: '{query}'...")

        # 1. Get initial content-based recommendations
        content_recs = self.content_recommender.get_recommendations(movie_title, top_n * 5) # Get more candidates initially
        if content_recs is None:
            return None

        print(f"   Found {len(content_recs)} initial content-based candidates.")

        # 2. Apply LLM-based filtering if a query is provided
        filtered_candidates = content_recs.copy()
        if query and self.gemini_interpreter and self.gemini_interpreter.model:
            print("   Interpreting query with Gemini...")
            interpretation = self.gemini_interpreter.interpret_query(query)
            if interpretation and interpretation.get('interpreted_filters'):
                filters = interpretation['interpreted_filters']
                print(f"   Applying interpreted filters: {filters}")

                # Apply genre filter
                if 'genres' in filters and filters['genres']:
                    genre_keywords = [g.lower() for g in filters['genres']]
                    filtered_candidates = filtered_candidates[
                        filtered_candidates['genres'].str.lower().apply(
                            lambda x: any(gk in x for gk in genre_keywords)
                        )
                    ]

                # Apply rating filter
                if 'min_rating' in filters:
                     filtered_candidates = filtered_candidates[
                         filtered_candidates['vote_average'] >= filters['min_rating']
                     ]
                if 'max_rating' in filters:
                     filtered_candidates = filtered_candidates[
                         filtered_candidates['vote_average'] <= filters['max_rating']
                     ]

                # Apply popularity filter
                if 'min_popularity' in filters:
                     filtered_candidates = filtered_candidates[
                         filtered_candidates['popularity'] >= filters['min_popularity']
                     ]
                if 'max_popularity' in filters:
                     filtered_candidates = filtered_candidates[
                         filtered_candidates['popularity'] <= filters['max_popularity']
                     ]

                print(f"   {len(filtered_candidates)} candidates remaining after filtering.")

            else:
                print("   Gemini interpretation failed or no filters generated. Proceeding with initial candidates.")
        else:
             print("   No query or Gemini interpreter not available. Proceeding with initial candidates.")


        if filtered_candidates.empty:
            print(" No candidates left after filtering.")
            return None


        # 3. Calculate hybrid scores for filtered candidates
        final_scores = []

        # Ensure feature columns align with df_modeling used for training
        valid_feature_cols = [col for col in self.feature_columns if col in self.df.columns]


        for idx, row in filtered_candidates.iterrows():
            movie_idx = row.name

            # Content similarity score (already in filtered_candidates)
            content_score = row['similarity_score']

            # Get features for predictive models from the original df (df_modeling)
            try:
                # Use df_modeling index to get features
                # Make sure the index exists after filtering/merging
                original_df_row = self.df.loc[self.df['title'] == row['title']]
                if original_df_row.empty:
                    print(f"Warning: Movie '{row['title']}' not found in original df_modeling for feature extraction.")
                    continue
                original_df_idx = original_df_row.index[0]


                # Select features for prediction
                movie_features = self.df.loc[original_df_idx, valid_feature_cols].values.reshape(1, -1)

                # Predict rating score (normalized 0-1)
                # Ensure the prediction is within 0-10 range before normalizing
                # The models were trained for classification, not regression.
                # Need to use a regression model for rating prediction or adjust logic.
                # For now, use the vote_average directly as a proxy for rating prediction score (scaled 0-1)
                predicted_rating = self.df.loc[original_df_idx, 'vote_average']
                rating_score = predicted_rating / 10.0 # Scale rating to 0-1


                # Use actual popularity (normalized 0-1)
                popularity = self.df.loc[original_df_idx, 'popularity'] # Use popularity from original df
                # Scale popularity score using min-max scaling based on the entire df
                max_popularity = self.df['popularity'].max()
                min_popularity = self.df['popularity'].min()
                popularity_score = (popularity - min_popularity) / (max_popularity - min_popularity) if (max_popularity - min_popularity) > 0 else 0
                # Ensure popularity score is between 0 and 1
                popularity_score = max(0, min(1, popularity_score))


                # Calculate hybrid score
                hybrid_score = (
                    content_weight * content_score +
                    rating_weight * rating_score +
                    popularity_weight * popularity_score
                )

                final_scores.append({
                    'movie_idx': movie_idx, # Keep original index for merging
                    'content_score': content_score,
                    'rating_score': rating_score,
                    'popularity_score': popularity_score,
                    'hybrid_score': hybrid_score
                })

            except Exception as e:
                print(f"Error processing movie {row['title']} for scores: {e}")
                continue

        if not final_scores:
            print(" No valid scores calculated for remaining candidates.")
            return None

        # Create scores dataframe
        score_df = pd.DataFrame(final_scores)

        # Merge scores with filtered candidates using the original index ('movie_idx')
        results_df = filtered_candidates.reset_index().merge(
            score_df[['movie_idx', 'content_score', 'rating_score', 'popularity_score', 'hybrid_score']],
            left_on='index', # Use the original index from filtered_candidates
            right_on='movie_idx'
        ).set_index('index') # Set the original index back

        # Sort by hybrid score and return top N
        hybrid_recommendations = results_df.nlargest(top_n, 'hybrid_score').reset_index(drop=True)


        # 4. Generate enhanced descriptions for the top recommendations
        if self.generative_enricher and self.generative_enricher.model:
            print("\n Generating enhanced descriptions for top recommendations...")
            # Initialize column outside the loop
            hybrid_recommendations['enhanced_description'] = None

            for index, row in hybrid_recommendations.iterrows():
                movie_title_to_enrich = row['title']
                enhanced_info = self.generative_enricher.enhance_descriptions(movie_title_to_enrich)
                if enhanced_info and 'enhanced_description' in enhanced_info:
                    # Ensure the description is a string and handle potential None/NaN
                    description_text = enhanced_info['enhanced_description']
                    if pd.notna(description_text):
                        hybrid_recommendations.loc[index, 'enhanced_description'] = description_text
                    else:
                         hybrid_recommendations.loc[index, 'enhanced_description'] = "Could not generate enhanced description."
                else:
                    hybrid_recommendations.loc[index, 'enhanced_description'] = "Could not generate enhanced description."
            print(" Enhanced descriptions generated.")
        else:
            print("\nâ—ï¸ Generative enricher not available. Skipping enhanced descriptions.")


        print(f"\n Hybrid recommendations generated with scores:")
        print(f"   Content weight: {content_weight:.2f}, Rating weight: {rating_weight:.2f}, Popularity weight: {popularity_weight:.2f}")

        return hybrid_recommendations[['title', 'genres', 'vote_average', 'popularity',
                                     'content_score', 'rating_score', 'popularity_score', 'hybrid_score', 'enhanced_description']]


# Initialize Hybrid Recommender
print("\n INITIALIZING HYBRID RECOMMENDER")
# Assuming content_recommender, rating_models, popularity_models, gemini_interpreter, and generative_enricher
# are already initialized from previous cells

# Pass the actual trained models instead of dictionaries
hybrid_recommender = HybridRecommender(
    df_modeling, # Use the df_modeling which includes all features
    content_recommender,
    xgb_model,  # Using XGBoost model for rating score (as a placeholder, ideally a regressor is needed)
    xgb_model,  # Using XGBoost model for popularity score (as a placeholder, ideally a regressor is needed)
    gemini_interpreter,
    generative_enricher
)

# Test hybrid recommender with user input
movie_input = input("Enter a movie title to get recommendations based on (e.g., Avatar): ")
query_input = input("Enter a natural language query for filtering (e.g., 'funny action movies with high ratings'): ")


print(f"\n TESTING HYBRID RECOMMENDATIONS FOR: '{movie_input}' with query '{query_input}'")
hybrid_recommendations = hybrid_recommender.get_hybrid_recommendations(movie_input, query=query_input, top_n=5)

if hybrid_recommendations is not None:
    print(f"\n Top 5 hybrid recommendations for '{movie_input}' filtered by '{query_input}':")
    # Display relevant columns, including the new enhanced description
    display_cols = ['title', 'genres', 'vote_average', 'hybrid_score', 'enhanced_description']
    # Ensure all display_cols exist in the result before printing
    display_cols = [col for col in display_cols if col in hybrid_recommendations.columns]

    # Print the dataframe, wrapping text in the enhanced description column
    for index, row in hybrid_recommendations[display_cols].iterrows():
        print("\n--- Recommendation ---")
        print(f"Title: {row['title']}")
        print(f"Genres: {row['genres']}")
        print(f"Vote Average: {row['vote_average']:.2f}")
        print(f"Hybrid Score: {row['hybrid_score']:.4f}")
        if 'enhanced_description' in row and pd.notna(row['enhanced_description']):
             print(f"   Description: {row['enhanced_description']}")
        else:
            print("   Description: Could not generate enhanced description.")
        print("----------------------")

else:
    print(f" Could not generate hybrid recommendations for '{movie_input}' with query '{query_input}'.")
````

````python
class CompleteRecommendationPipeline:
    def __init__(self, df, hybrid_recommender, query_interpreter, generative_enricher):
        self.df = df
        self.hybrid_recommender = hybrid_recommender
        self.query_interpreter = query_interpreter
        self.generative_enricher = generative_enricher

    def process_query(self, query, top_n=5):
        """
        Complete pipeline: Query -> Interpretation -> Recommendation -> Enhancement
        """
        print(f"\n{'='*70}")
        print(f" PROCESSING QUERY: '{query}'")
        print(f"{'='*70}")

        # Step 1: Query Interpretation
        print("\n1.  QUERY INTERPRETATION")
        interpretation = self.query_interpreter.interpret_query(query)

        if not interpretation:
            print("    Query interpretation failed.")
            return None

        print("    Interpretation Results:")
        print(f"      â€¢ Genres: {interpretation.get('genres')}")
        print(f"      â€¢ Mood: {interpretation.get('mood')}")
        print(f"      â€¢ Rating Preference: {interpretation.get('rating_preference')}")
        print(f"      â€¢ Source: {interpretation['source']}")
        print(f"      â€¢ Interpreted Filters: {interpretation.get('interpreted_filters')}")


        # Step 2: Find base movie and generate recommendations
        print(f"\n2.  GENERATING RECOMMENDATIONS")
        base_movie = self._find_base_movie(interpretation)

        if not base_movie:
            print("    Could not find suitable base movie based on interpretation.")
            # Fallback: If no base movie found from interpretation, maybe use a popular movie?
            # For now, we'll just return None. You might want to add fallback logic here.
            return None

        print(f"   â€¢ Using base movie: '{base_movie}'")
        # Pass the original query to the hybrid recommender for potential additional filtering
        recommendations = self.hybrid_recommender.get_hybrid_recommendations(base_movie, query=query, top_n=top_n)


        if recommendations is None or len(recommendations) == 0:
            print("    No recommendations generated")
            return None

        print(f"   â€¢ Generated {len(recommendations)} recommendations")

        # Step 3: Display final results (Enhancement is done within hybrid_recommender)
        print(f"\n4.  FINAL RECOMMENDATIONS")
        self._display_results(recommendations, interpretation)

        return recommendations

    def _find_base_movie(self, interpretation):
        """Find optimal base movie for recommendations based on interpretation filters."""
        filters = interpretation.get('interpreted_filters', {})
        filtered_df = self.df.copy()

        print(f"   â€¢ Finding base movie with filters: {filters}")

        # Apply genre filters
        if 'genres' in filters and filters['genres']:
            genre_keywords = [g.lower() for g in filters['genres']]
            # Filter movies where the genres column contains ANY of the genre keywords
            filtered_df = filtered_df[
                filtered_df['genres'].str.lower().apply(
                    lambda x: any(gk in x for gk in genre_keywords)
                )
            ]
            print(f"     - After genre filter: {len(filtered_df)} movies")


        # Apply rating filter
        if 'min_rating' in filters:
             filtered_df = filtered_df[
                 filtered_df['vote_average'] >= filters['min_rating']
             ]
             print(f"     - After min_rating filter: {len(filtered_df)} movies")

        if 'max_rating' in filters:
             filtered_df = filtered_df[
                 filtered_df['vote_average'] <= filters['max_rating']
             ]
             print(f"     - After max_rating filter: {len(filtered_df)} movies")

        # Apply popularity filter
        if 'min_popularity' in filters:
             filtered_df = filtered_df[
                 filtered_df['popularity'] >= filters['min_popularity']
             ]
             print(f"     - After min_popularity filter: {len(filtered_df)} movies")

        if 'max_popularity' in filters:
             filtered_df = filtered_df[
                 filtered_df['popularity'] <= filters['max_popularity']
             ]
             print(f"     - After max_popularity filter: {len(filtered_df)} movies")


        # Select a base movie from the filtered results
        if not filtered_df.empty:
            # Option 1: Select the most popular movie among the filtered results
            base_movie_title = filtered_df.sort_values(by='popularity', ascending=False).iloc[0]['title']
            return base_movie_title
            # Option 2: Select the highest rated movie among the filtered results
            # base_movie_title = filtered_df.sort_values(by='vote_average', ascending=False).iloc[0]['title']
            # return base_movie_title
            # Option 3: Select a random movie from the filtered results
            # base_movie_title = filtered_df.sample(1).iloc[0]['title']
            # return base_movie_title
        else:
            return None # No movies match the filters

    def _display_results(self, recommendations, interpretation):
        """Display the final recommendations in a user-friendly format."""
        print(f"\nRecommendations based on your query: '{interpretation['original_query']}'")
        print("-" * (len(interpretation['original_query']) + 30))

        if recommendations is not None and not recommendations.empty:
            for index, row in recommendations.iterrows():
                print(f"\n{index + 1}. **{row['title']}**")
                print(f"   Genres: {row['genres']}")
                print(f"   Rating: {row['vote_average']:.2f}")
                print(f"   Popularity: {row['popularity']:.2f}")
                print(f"   Hybrid Score: {row['hybrid_score']:.4f}")
                if 'enhanced_description' in row and pd.notna(row['enhanced_description']):
                     print(f"   Description: {row['enhanced_description']}")
                # You could add more details or an explanation of why it was recommended here
                # based on the interpretation and scores.
        else:
            print("No recommendations found matching your criteria.")

# Example Usage (after initializing other components)
# Assuming content_recommender, rating_models, popularity_models, gemini_interpreter, and generative_enricher are initialized

# Initialize the complete pipeline
complete_pipeline = CompleteRecommendationPipeline(
    df_modeling,
    hybrid_recommender,
    gemini_interpreter,
    generative_enricher
)

# Get user query
user_query = input("Enter your movie recommendation request (e.g., 'funny action movies with high ratings'): ")

# Process the query and get recommendations
final_recommendations = complete_pipeline.process_query(user_query)

# Note: The example usage is commented out to avoid running automatically.
# You can uncomment and run this cell after initializing all required components.
````
